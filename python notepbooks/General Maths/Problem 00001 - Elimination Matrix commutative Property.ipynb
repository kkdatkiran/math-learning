```python
import numpy as np

def elimination_matrix(size, i, j, value):
    mat = np.identity(size)
    mat[i][j] = value
    return mat

def prime_number(min, max):
    return np.random.permutation([x for x in range(min, max) if not [t for t in range(2, x) if not x % t]])[0]
```

```python
size = 50
i = 5
j = 4
eij = elimination_matrix(size, i - 1, j - 1, prime_number(2300, 5000))

for p in range(1, size):
    line = f"Row : {p+1} : \n"
    for q in range(0, p):
        epq = elimination_matrix(size, p, q, prime_number(5300, 8000))
        epqij = np.matmul(epq, eij)
        eijpq = np.matmul(eij, epq)
        if np.array_equal(epqij, eijpq):
            line = line + f"[{p+1},{q+1}] "
    print(line)
```
