```python
import numpy as np
import math
```

```python
def negate(func):
    return lambda x: -func(x)

def one(value):
    return 1

def zero(value):
    return 0

def square(func):
    return lambda x: func(x) * func(x)

def inverse(func):
    return lambda x: 1/func(x)

def __gcd(a, b):
  
    if (a == 0 or b == 0): return 0
    if (a == b): return a
    if (a > b): 
        return __gcd(a - b, b)            
    return __gcd(a, b - a)

def coprimes(a, b):
    return __gcd(a, b) == 1

matrices = [
    [50, [math.cos, negate(math.sin), math.sin, math.cos], "[cos, -sin, sin, cos]"],
    [50, [math.cos, math.sin, negate(math.sin), math.cos], "[cos, sin, -sin, cos]"],
    [50, [negate(math.cos), negate(math.sin), math.sin, negate(math.cos)], "[-cos, -sin, sin, -cos]"],
    [50, [negate(math.cos), math.sin, negate(math.sin), negate(math.cos)], "[-cos, sin, -sin, -cos]"],
    [50, [inverse(math.sin), inverse(negate(math.cos)), inverse(math.cos), inverse(math.sin)], "[cosec, -sec, sec, cosec]"],
    [50, [inverse(math.sin), inverse(math.cos), inverse(negate(math.cos)), inverse(math.sin)], "[cosec, sec, -sec, cosec]"],
    [50, [inverse(negate(math.sin)), inverse(negate(math.cos)), inverse(math.cos), inverse(negate(math.sin))], "[-cosec, -sec, sec, -cosec]"],
    [50, [inverse(negate(math.sin)), inverse(math.cos), inverse(negate(math.cos)), inverse(negate(math.sin))], "[-cosec, sec, -sec, -cosec]"],
]

for eachSet in matrices:
    print(f"Set {eachSet[2]} : ")
    for multiple in [1,2,3,4]:
        divi = eachSet[0] * multiple
        coprime = 1
        count = 0
        
        while count < 4:
            # if not coprimes(coprime, divi):
            #     coprime += 1
            #     continue
            count += 1

            a = np.array([
                [eachSet[1][0]((math.pi * coprime)/divi), eachSet[1][1]((math.pi * coprime)/divi)],
                [eachSet[1][2]((math.pi * coprime)/divi), eachSet[1][3]((math.pi * coprime)/divi)]])
            b = np.copy(a)
            
            for i in range(1, 51):
                vals = np.linalg.eigvals(b)
                if not all([val.imag != 0 for val in vals]):
                    print(f"Checking for ({coprime} * pi)/{divi} : ")        
                    print(f"Eigen values are not complex for {i}th power but from 1 to {i-1}th power is complex")
                    break
                b = np.matmul(b, a)
            coprime += 1
                
```
